# 📘 Binomio Cuadrado Perfecto - Representación Geométrica 📊

En este notebook explicaremos el concepto del **binomio cuadrado perfecto** y su representación geométrica.

### ¿Qué es el binomio cuadrado perfecto?
Es una identidad algebraica de la forma:

$$
(a + b)^2 = a^2 + 2ab + b^2
$$

Nos dice que el cuadrado de la suma de dos términos es igual a la suma de los cuadrados individuales más el doble del producto de ambos.

En este notebook veremos su demostración algebraica y una visualización geométrica con Python. 🚀

```python
import numpy as np
import plotly.graph_objects as go

# Función para representar gráficamente el binomio cuadrado perfecto en 3D
def binomio_cuadrado_3d(a, b):
    X = np.linspace(-a - b, a + b, 50)
    Y = np.linspace(-a - b, a + b, 50)
    X, Y = np.meshgrid(X, Y)
    Z = X**2 + 2*X*Y + Y**2  # Expansión del binomio cuadrado perfecto

    # Crear la figura interactiva con Plotly
    fig = go.Figure()
    fig.add_trace(go.Surface(z=Z, x=X, y=Y, colorscale='viridis'))

    fig.update_layout(
        title=f'Representación 3D del Binomio Cuadrado Perfecto: ({a} + {b})²',
        scene=dict(
            xaxis_title='Valor de X',
            yaxis_title='Valor de Y',
            zaxis_title='X² + 2XY + Y²'
        )
    )
    fig.show()

# Parámetros iniciales
a = 2
b = 2

# Llamar a la función para visualizar el gráfico 3D
binomio_cuadrado_3d(a, b)
```

## 📊 **Resultados**

1️⃣ Se logró visualizar de manera interactiva la representación 3D del **Binomio Cuadrado Perfecto**: $(a + b)^2 = a^2 + 2ab + b^2$
   mediante una superficie en el espacio tridimensional.

2️⃣ La gráfica muestra cómo varía la expresión $( X^2 + 2XY + Y^2 )$ en función de ( X ) y ( Y ), reflejando la expansión del binomio cuadrado perfecto.

3️⃣ Al modificar los valores de ( a ) y ( b ), se observa un cambio en la escala de la superficie, confirmando la dependencia de los coeficientes en la expresión cuadrática.

📌 **Interpretación Geométrica**  
- La ecuación **$( (a + b)^2 = a^2 + 2ab + b^2 )$** tiene una representación geométrica clara en el espacio tridimensional.  
- La superficie generada por la función **$( X^2 + 2XY + Y^2 )$** es una **parábola hiperbólica**, lo que muestra la naturaleza cuadrática del binomio.
