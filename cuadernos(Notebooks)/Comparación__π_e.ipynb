# **Comparación entre $π^e$ y $e^π$**

## 📖 **Introducción**

Los números irracionales $(\pi)$ y $(e)$ son fundamentales en matemáticas. En este estudio, analizaremos cuál de los valores $(\pi^e)$ o $(e^\pi)$ es mayor utilizando herramientas matemáticas como:

- Comparación numérica directa.
- Análisis de funciones relacionadas.
- Cálculo de máximos y mínimos mediante derivadas.

Este notebook proporciona una **explicación visual e interactiva** para comprender mejor la relación entre estas expresiones.


```python
# Importamos las bibliotecas necesarias
import numpy as np
import sympy as sp
import plotly.graph_objects as go

# Definimos los valores de π y e
pi_val = np.pi
e_val = np.e

# Comparación directa
pi_e = pi_val**e_val
e_pi = e_val**pi_val
resultado = "π^e < e^π" if pi_e < e_pi else "π^e > e^π"

# Definimos las funciones matemáticas para el análisis
x = sp.symbols('x')
f = sp.exp(x) - 1 - x   # f(x) = e^x - 1 - x
g = sp.ln(x) / x        # g(x) = ln(x) / x
h = x**(1/x)            # h(x) = x^(1/x)

# Generamos valores para la gráfica
x_vals = np.linspace(0.1, 5, 100)

# Convertimos valores simbólicos a float
f_vals = [float(f.subs(x, val)) for val in x_vals]
g_vals = [float(g.subs(x, val)) for val in x_vals]
h_vals = [float(h.subs(x, val)) for val in x_vals]

# Puntos especiales en h(x) para x = e y x = π
h_e = float(h.subs(x, e_val))
h_pi = float(h.subs(x, pi_val))

# Visualización interactiva con Plotly
fig = go.Figure()

# Funciones
fig.add_trace(go.Scatter(x=x_vals, y=f_vals, mode='lines', name='f(x) = e^x - 1 - x', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=x_vals, y=g_vals, mode='lines', name='g(x) = ln(x)/x', line=dict(color='green')))
fig.add_trace(go.Scatter(x=x_vals, y=h_vals, mode='lines', name='h(x) = x^(1/x)', line=dict(color='red', width=3)))

# Puntos clave en h(x)
fig.add_trace(go.Scatter(
    x=[e_val], y=[h_e], mode='markers', name='h(e)',
    marker=dict(color='red', size=10, symbol='circle')
))
fig.add_trace(go.Scatter(
    x=[pi_val], y=[h_pi], mode='markers', name='h(π)',
    marker=dict(color='red', size=10, symbol='x')
))

# Líneas punteadas para destacar los valores
fig.add_trace(go.Scatter(x=[e_val, e_val], y=[0, h_e], mode='lines', line=dict(dash='dot', color='gray'), showlegend=False))
fig.add_trace(go.Scatter(x=[pi_val, pi_val], y=[0, h_pi], mode='lines', line=dict(dash='dot', color='gray'), showlegend=False))

# Configuración del gráfico
fig.update_layout(
    title=f"📊 Comparación de funciones para π^e y e^π: {resultado}",
    xaxis_title="x",
    yaxis_title="f(x), g(x), h(x)",
    legend_title="Funciones",
    margin=dict(l=20, r=20, t=40, b=20),
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True)
)

# Mostramos la gráfica interactiva
fig.show()
```

## 📢 **Resultados**
La gráfica muestra el comportamiento de tres funciones matemáticas que se utilizan en demostraciones alternativas de la desigualdad:  

$$
\pi^e < e^\pi
$$

## Las funciones representadas son:  

### 1️⃣ $ f(x) = e^x - 1 - x)$  
- Representa la diferencia entre la exponencial y la recta ( y = x + 1 ).  
- Se usa para analizar convexidad y crecimiento exponencial.  

### 2️⃣ $( g(x) = {ln x}{x} )$  
- Se emplea para estudiar el máximo de la función \( x^{1/x} \), clave en la comparación de potencias.  

### 3️⃣ $(h(x) = x^{1/x})$
- Permite visualizar en qué punto la base ( x ) elevada a su propio recíproco alcanza su máximo.  
- Se usa para comparar ( $\pi^e$ ) y $( e^\pi)$, ya que este máximo ocurre en $( x = e )$.  

---

### 🔎 ¿Qué observamos en la gráfica?  

**La función $( h(x) = x^{1/x})$ tiene un máximo en $(x = e)$**  
- Esto significa que la base óptima para maximizar potencias de la forma $( x^{1/x} ) es ( e )$.  
- Como $(\pi\ approx\ 3.14 )$ es mayor que $(e \approx 2.71 )$, se encuentra en la **parte decreciente** de $( h(x))$, lo que sugiere que:  
  
$$
pi^e < e^\pi
$$

**La función $( g(x) = \ln x / x )$ tiene un máximo en \( x = e \)**  
- Este comportamiento también respalda que $( e^\pi )$ sea mayor que $( \pi^e )$.  

**La función $( f(x) = e^x - 1 - x )$ crece rápidamente**  
- Indica que la función exponencial supera linealmente a \( x + 1 \), lo cual es clave en ciertas demostraciones de la desigualdad.  


La gráfica confirma visualmente la desigualdad $( \pi^e < e^\pi )$, mostrando que la función $( x^{1/x} )$ alcanza su máximo en $( x = e )$, y que valores mayores a $( e )$ llevan a disminuciones en la función, favoreciendo a $( e^\pi )$.  

📌 **Intuición matemática:** La base óptima para exponenciaciones de la forma $( x^{1/x} )$ es $( e )$, lo que explica por qué $( e^\pi )$ es mayor que $( \pi^e )$. 🚀  






A través de este análisis, obtenemos los valores:

- $( pi^e ) ≈ {pi_e:.6f}$
- $( e^\pi ) ≈ {e_pi:.6f}$

Observamos que **$( e^\pi)$** es mayor que **$( \pi^e )$** Este resultado puede demostrarse utilizando propiedades del cálculo y análisis de funciones.

- El análisis de $( g(x) = \ln(x)/x)$ nos muestra su decrecimiento en \( x > 1 \).
