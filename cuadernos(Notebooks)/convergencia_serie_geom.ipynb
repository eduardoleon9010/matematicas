# **Animación de la Convergencia de una Serie Geométrica**

### **1. Introducción**
Las series geométricas desempeñan un papel fundamental en matemáticas, especialmente en análisis y cálculo. Una serie geométrica se define como:

$$
S_n = a + ar + ar^2 + ar^3 + \dots + ar^{n-1}
$$

donde:
- $( a )$ es el primer término de la serie.
- \( r \) es la razón común entre los términos.
- \( n \) representa el número de términos.

Si el valor absoluto de la razón \( r \) es menor que 1, la serie converge a un valor finito dado por la fórmula:

$$
S = \frac{a}{1 - r}, \quad \text{si } |r| < 1.
$$

Este proyecto busca demostrar visualmente la convergencia de una serie geométrica mediante una animación en Python usando **Matplotlib**.

### **2. Objetivo**
Los objetivos de esta animación son:
- Visualizar gráficamente la suma parcial de la serie geométrica y su convergencia.
- Ilustrar la aplicación de la fórmula de suma infinita $( S = \frac{a}{1 - r} )$.
- Utilizar **Matplotlib** para crear una animación interactiva de la suma de las series.

### **3. Definición del problema**
La idea principal es representar gráficamente las sumas parciales \( S_n \) de la serie geométrica y compararlas con el valor límite \( S \). La animación mostrará el crecimiento de las sumas parciales hasta su estabilización, destacando visualmente la convergencia.



```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Parámetros de la serie geométrica
a = 1  # Primer término
r = 0.5  # Razón geométrica (|r| < 1 para convergencia)
n_terms = 30  # Número de términos a visualizar

# Cálculo de las sumas parciales
n_values = np.arange(n_terms)
suma_parcial = a * (1 - r**(n_values + 1)) / (1 - r)
suma_infinita = a / (1 - r)  # Valor límite de la serie

# Configuración de la figura y ejes
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlim(-1, n_terms)
ax.set_ylim(0, suma_infinita + 0.5)
ax.set_xlabel("Número de términos (n)")
ax.set_ylabel("Suma Parcial")
ax.set_title("Convergencia de una Serie Geométrica")

# Línea de referencia del valor límite
ax.axhline(y=suma_infinita, color='r', linestyle='dashed', label=f"Suma infinita = {suma_infinita:.2f}")

# Línea y puntos de la serie
line, = ax.plot([], [], 'bo-', label="Suma Parcial")
point, = ax.plot([], [], 'ro', markersize=8)  # Punto móvil

# Función de actualización de la animación
def update(i):
    x_data = n_values[:i+1]  # Asegurar que es una lista de valores
    y_data = suma_parcial[:i+1]

    line.set_data(x_data, y_data)  # Actualizar la serie
    point.set_data([x_data[-1]], [y_data[-1]])  # Punto en el último término
    return line, point

# Crear la animación
ani = animation.FuncAnimation(fig, update, frames=n_terms, interval=500, blit=True)

# Mostrar la animación en Google Colab
HTML(ani.to_jshtml())

```

## **4. Resultados**
La implementación generó una animación interactiva que muestra la suma parcial de la serie geométrica a lo largo del tiempo. Se utilizaron puntos y líneas para ilustrar la evolución de \( S_n \) y su aproximación al valor límite \( S \). La animación se ejecutó con éxito en **Google Colab** utilizando `matplotlib.animation`.
