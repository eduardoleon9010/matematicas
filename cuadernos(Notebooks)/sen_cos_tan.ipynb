# Propiedades de las funciones seno, coseno y tangente

## 📖 Introducción  
Las funciones trigonométricas seno, coseno y tangente son fundamentales en el estudio de la geometría y el análisis matemático. Estas funciones describen relaciones entre los ángulos y los lados de un triángulo rectángulo y tienen aplicaciones en diversas áreas como la física, la ingeniería y la computación.  

Para comprender mejor estas funciones, utilizamos el **círculo unitario**, donde un ángulo **θ** se mide en sentido contrario a las agujas del reloj desde el eje **x positivo**. A partir de este círculo, podemos derivar importantes propiedades que rigen el comportamiento del seno, coseno y tangente.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'


# Definir el ángulo θ en radianes (45° como ejemplo)
theta = np.pi / 4

# Círculo unitario
circle = plt.Circle((0, 0), 1, color='blue', fill=False, linestyle='dashed')

# Puntos clave
x = np.cos(theta)  # Coseno de θ
y = np.sin(theta)  # Seno de θ
tan_x = 1  # Línea tangente toca x = 1
tan_y = np.tan(theta)  # Valor de la tangente

# Crear la figura
fig, ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal')
ax.set_xticks([-1, 0, 1])
ax.set_yticks([-1, 0, 1])
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Agregar el círculo unitario
ax.add_patch(circle)

# Dibujar el radio del círculo (hipotenusa)
ax.plot([0, x], [0, y], 'r-', linewidth=2, label=r'$r = 1$')

# Dibujar los catetos del triángulo
ax.plot([x, x], [0, y], 'g--', label=r'$\sin(\theta)$')
ax.plot([0, x], [y, y], 'b--', label=r'$\cos(\theta)$')

# Línea tangente
ax.plot([1, 1], [0, tan_y], 'purple', linestyle='dotted', label=r'$\tan(\theta)$')

# Etiquetas de puntos clave
ax.text(x/2, y/2, r'$\theta$', fontsize=12, color='red', verticalalignment='bottom')
ax.text(x, -0.1, r'$\cos(\theta)$', fontsize=12, color='blue', horizontalalignment='center')
ax.text(-0.1, y, r'$\sin(\theta)$', fontsize=12, color='green', verticalalignment='center')
ax.text(1.05, tan_y/2, r'$\tan(\theta)$', fontsize=12, color='purple')

# Mostrar leyenda
ax.legend(loc='upper right')

# Mostrar la gráfica
plt.title("📌 Círculo Unitario y Funciones Trigonométricas")
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

# Asegurar que Plotly renderiza en Google Colab
pio.renderers.default = "colab"

# Función para generar el gráfico
def crear_grafico(theta_deg=45):
    theta = np.radians(theta_deg)
    x, y = np.cos(theta), np.sin(theta)
    tan_y = np.tan(theta) if np.abs(np.cos(theta)) > 1e-2 else None

    # Crear figura
    fig = go.Figure()

    # Círculo unitario
    t = np.linspace(0, 2*np.pi, 100)
    fig.add_trace(go.Scatter(x=np.cos(t), y=np.sin(t), mode="lines", name="Círculo Unitario"))

    # Triángulo
    fig.add_trace(go.Scatter(x=[0, x, x], y=[0, 0, y], mode="lines", line=dict(color="red"), name="Triángulo"))

    # Radio
    fig.add_trace(go.Scatter(x=[0, x], y=[0, y], mode="lines", line=dict(color="black", width=2), name="r = 1"))

    # Punto en el círculo
    fig.add_trace(go.Scatter(x=[x], y=[y], mode="markers", marker=dict(color="black", size=8), name=f"P({x:.2f}, {y:.2f})"))

    # Líneas seno y coseno
    fig.add_trace(go.Scatter(x=[x, x], y=[0, y], mode="lines", line=dict(color="green", dash="dash"), name="sen(θ)"))
    fig.add_trace(go.Scatter(x=[0, x], y=[y, y], mode="lines", line=dict(color="blue", dash="dash"), name="cos(θ)"))

    # Línea tangente
    if tan_y is not None and abs(tan_y) < 5:
        fig.add_trace(go.Scatter(x=[1, 1], y=[0, tan_y], mode="lines", line=dict(color="purple", dash="dot"), name="tan(θ)"))

    # Layout
    fig.update_layout(
        title=f"Círculo Unitario - Ángulo θ = {theta_deg}°",
        xaxis=dict(range=[-1.2, 1.2], zeroline=True),
        yaxis=dict(range=[-1.2, 1.2], zeroline=True),
        showlegend=True,
        width=600, height=600
    )

    fig.show()  # Mostrar la gráfica

# Llamar la función manualmente para ver el gráfico inicial
crear_grafico(45)
```


## 📊 Resultados  
Al representar un ángulo **θ** en el círculo unitario:  
✅ **El coseno (cos θ)** corresponde a la coordenada **x** del punto sobre el círculo.  
✅ **El seno (sen θ)** corresponde a la coordenada **y** del punto.  
✅ **La tangente (tan θ)** se define como la razón entre el seno y el coseno, es decir, **$tan(θ) = sen(θ) / cos(θ)$**.  

### **Otras propiedades clave**  
📌 **Relaciones fundamentales**  
- $( \sin^2(\theta) + \cos^2(\theta) = 1 ) $
- $( \tan(\theta) = \frac{\sin(\theta)}{\cos(\theta)} )$  
- $( \tan(90^\circ) ) no\ está\ definida\ porque ( \cos(90^\circ) = 0 )$.  

📌 **Simetrías en el círculo unitario**  
- $( \sin(-\theta) = -\sin(\theta) ) (funcion\ impar)$  
- $( \cos(-\theta) = \cos(\theta) ) (funcion\ par)$
- $( \tan(-\theta) = -\tan(\theta) ) $
