## **Introducción**  
Los sistemas de coordenadas son esenciales en matemáticas y física para representar posiciones en el espacio. Dos de los sistemas más utilizados son:  
- **Coordenadas cartesianas**, que emplean ejes ortogonales para ubicar puntos en el plano o en el espacio tridimensional.  
- **Coordenadas polares**, donde cada punto es determinado por un radio y un ángulo con respecto a un origen de referencia.  

Este cuaderno de Google Colab proporciona una explicación detallada de ambos sistemas, con ejemplos visuales e interactivos para facilitar su comprensión.

```python
# Importación de Librerías
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

### 1. Representación del plano Cartesiano


```python
def plano_cartesiano():
    fig, ax = plt.subplots(figsize=(6,6))
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.grid(True, linestyle='--', linewidth=0.5)
    ax.set_title("Plano Cartesiano")
    plt.show()

plano_cartesiano()
```

### 2. Representación del *sistema* Polar

```python
def plano_polar():
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(6,6))
    ax.set_title("Coordenadas Polares")
    plt.show()

plano_polar()
```

### 3. Conversión entre Coordenadas


```python
def cartesianas_a_polares(x, y):
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    return r, np.degrees(theta)

def polares_a_cartesianas(r, theta):
    theta_rad = np.radians(theta)
    x = r * np.cos(theta_rad)
    y = r * np.sin(theta_rad)
    return x, y
```

### 4. Visualización interactiva


```python
def graficar_conversion(x=3, y=4):
    r, theta = cartesianas_a_polares(x, y)
    x_polar, y_polar = polares_a_cartesianas(r, theta)

    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    # Gráfico cartesiano
    ax[0].axhline(0, color='black', linewidth=1)
    ax[0].axvline(0, color='black', linewidth=1)
    ax[0].grid(True, linestyle='--', linewidth=0.5)
    ax[0].scatter(x, y, color='red', label=f"({x}, {y})")
    ax[0].legend()
    ax[0].set_title("Coordenadas Cartesianas")

    # Gráfico polar
    ax[1] = plt.subplot(1, 2, 2, projection='polar')
    ax[1].plot([0, np.radians(theta)], [0, r], marker='o', color='blue')
    ax[1].set_title("Coordenadas Polares")

    plt.show()

interact(graficar_conversion, x=(-10,10,1), y=(-10,10,1));
```
