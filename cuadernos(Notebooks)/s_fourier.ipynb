## 📝 **Introducción**
Jean-Baptiste Joseph Fourier fue un matemático y físico francés que revolucionó la forma en que entendemos los fenómenos periódicos. Desarrolló la **Serie de Fourier**, una herramienta matemática que permite representar una función periódica como la suma infinita de funciones senoidales (senos y cosenos).

Las **Series de Fourier** tienen aplicaciones en muchas áreas como:
- 📡 **Procesamiento de señales** (sonido, imágenes, video).
- 🔥 **Fenómenos de transferencia de calor**.
- 🎼 **Compresión de audio y video**.
- ⚡ **Electromagnetismo y circuitos eléctricos**.

## 🎯 **Objetivo**
El objetivo de este cuaderno es:
1. Explicar la teoría de la **Serie de Fourier**.
2. Mostrar visualmente cómo una onda cuadrada se puede construir con una suma de senos.
3. Permitir la **exploración interactiva** para ver el efecto de agregar más términos a la serie.

## 📚 **Fundamentos matemáticos**
La serie de Fourier de una función periódica $( f(x) )$ se expresa como:

$$
f(x) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n\omega x) + b_n \sin(n\omega x) \right)
$$

Para la **onda cuadrada**, la aproximación con \( N \) términos está dada por:

$$
f_N(x) = \sum_{k=1,3,5,\dots}^{N} \frac{4}{k\pi} \sin(kx)
$$

donde solo aparecen los términos impares. A medida que aumentamos \( N \), la aproximación mejora.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from ipywidgets import interact, IntSlider
from IPython.display import HTML

# Función para calcular la serie de Fourier de una función periódica
def fourier_series(x, n_terms):
    """Aproxima una función periódica con una serie de Fourier."""
    y = np.zeros_like(x)
    for n in range(1, n_terms + 1, 2):  # Solo términos impares para onda cuadrada
        y += (4 / (n * np.pi)) * np.sin(n * x)
    return y

# Crear figura para la animación
fig, ax = plt.subplots(figsize=(8, 4))
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)
ax.set_title("Aproximación de una Onda Cuadrada con Serie de Fourier")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid()

x = np.linspace(0, 2 * np.pi, 400)
line, = ax.plot(x, fourier_series(x, 1), 'b', label="Serie de Fourier")
ax.legend()

# Función de actualización de la animación
def update(n):
    line.set_ydata(fourier_series(x, n))
    return line,

# Crear animación
ani = animation.FuncAnimation(fig, update, frames=np.arange(1, 20, 2), interval=500, blit=True)
HTML(ani.to_jshtml())

# Agregar interactividad con sliders
def interactive_fourier(n_terms):
    plt.figure(figsize=(8, 4))
    plt.plot(x, fourier_series(x, n_terms), 'b', label=f"{n_terms} términos")
    plt.title("Serie de Fourier Interactiva")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.legend()
    plt.show()

interact(interactive_fourier, n_terms=IntSlider(min=1, max=50, step=1, value=5));
```

## **Resultados**
Tras el desarrollo y ejecucion del script interactivo y observamos:

-> Cómo la aproximación mejora al agregar más términos.

-> El **efecto de la oscilación de Gibbs**, que aparece cerca de los saltos de la onda cuadrada.

-> Cómo la **Serie de Fourier** nos ayuda a representar funciones complicadas con solo senos y cosenos.




## **Conclusiones**

-> A medida que agregamos más términos, la aproximación de la onda cuadrada mejora.
